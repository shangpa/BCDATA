### **Pandas 자료형: 시리즈(Series)**
##### > ***1차원 배열***로 여러 값을 나열한 자료 구조

```python
## 시리즈(Series) 형: 생성, index 강제 부여
import pandas as pd

a = pd.Series([11, 22, 33, 44], index = [1, 2, 3, 4])  #index 강제 부여
a
```

```python
## 시리즈(Series) 형: 생성 index 자동 부여
import pandas as pd

a = pd.Series([11, 22, 33, 44])  #index 자동 부여
a
```

### **Pandas 자료형: 데이터 프레임(Data Frame)**

##### > 행과 열 값들은 딕셔너리 구조를 활용하여 초기화 한다.

##### > pandas 패키지를 통해 사용할 수 있다.

### **데이터 추출 : loc[] 이용**
##### > loc[]는 데이터프레임에서 특정 행과 열을 위치 값(location)으로 선택하는 인덱싱(indexing) 방법 중 하나
##### > Index는 행을 식별하는 Key 역할을 함.
##### > ***Indexing***은 Index를 이용해 데이터를 추출하는 방법 > 결과적으로 ***행을 추출***
##### > loc[***row_index, col_index***]로 추출
##### > loc[***x***:***y***] 숫자 index가 x 이상 ~ y 이하 행을 추출

```python
## [Data Frame] 생성
import pandas as pd

df = pd.DataFrame({'Class' : [1, 1, 2],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]}, 
                  index =  ['Kims', 'Lees', 'Parks'])  #index 강제 부여
df
```

```python
## [Data Frame] 데이터 추출: loc[] > 행, 열 추출
# loc[row_index, col_index]
result = df.loc['Kims', 'Eng']  #결과는 Series 자료형
print(type(result))
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 행 (Series 자료형) 추출
# loc[row_index, col_index]
result = df.loc['Kims']  #결과는 Series 자료형
print(type(result))
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 행 (Data Frame 자료형) 추출
result = df.loc[['Kims', 'Lees']]  #결과는 Data Frame 자료형
print(type(result))
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 행 (Data Frame 자료형) 추출
result = df.loc[['Kims']]  #결과는 Data Frame 자료형
print(type(result))
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 복수 행 추출
result = df.loc[:]   #전체 행 선택, 결과는 Data Frame 자료형
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 행 (Data Frame 자료형) 추출
result = df.loc['Kims':'Lees']  #결과는 Data Frame 자료형
print(type(result))
result
```

```python
## [Data Frame] 데이터 추출: loc[] > 복수 행, 선택 열 추출
# loc[row_index, col_index]
result = df.loc[['Kims', 'Lees'], 'Mat']    #결과는 Series 자료형
result = df.loc[['Kims', 'Lees'], ['Mat']]  #결과는 Data Frame 자료형
result
```

### **데이터 추출 : iloc[] 이용**
##### > iloc[]는 데이터 추출 전에 ***행과 열에 임시로 숫자 index를 지정***하여 사용하는 인덱싱 방법
##### > loc[]와 달리 조건부 추출이 불가능
##### > iloc[***x***:***y***] ***x 이상 ~ y 미만*** 행을 추출 >> loc[***x***:***y***] 숫자 index가 ***x 이상 ~ y 이하*** 행을 추출  

```python
## [Data Frame] 생성
import pandas as pd

df = pd.DataFrame({ 'Name' : ['Kims', 'Lees', 'Parks'],
                   'Class' : [1, 1, 2],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]})  #index 자동 부여
df
```

```python
## [Data Frame] 데이터 추출: iloc[] > 행 추출
result = df.iloc[0]      #1개 데이터이므로 Series 형으로 추출
result = df.iloc[[0]]    #Data Frame 형으로 추출
result = df.iloc[:]      #복수개 데이터이므로 Data Frame 형으로 추출
result = df.iloc[1:3]    #복수개 데이터이므로 Data Frame 형으로 추출
result
```

```python
## [Data Frame] 데이터 추출: iloc[] > 행 추출
result = df.iloc[[0, 2]]    #복수개 데이터이므로 Data Frame 형으로 추출
result
```

#### **Index 번호로 열 추출**

```python
## [Data Frame] 데이터 추출: iloc[] > 행, 열 추출
result = df.iloc[[0, 2], [1]]       #열 선택도 열 순서 번호(index)로 명시
result = df.iloc[[0, 2], [1, 0]]    #열 선택도 열 순서 번호(index)로 명시
result
```

### **(알아 두면 좋아요) 열 변수를 index로 전환하기**
##### index를 열 변수로 전환도 가능(단, index에 변수명이 있을 때만 가능)
##### **.set_index()** : 특정 열을 index로 사용
###### > key : 인덱스로 사용하려는 열
###### > drop : 인덱스 제거 (default는 True)
###### > append : 자동 인덱스 추가 (default는 False), 멀티 인덱스 생성 시 사용
###### > inplace : False이면 자신은 안 바꾸고 변경 결과만 반환 (default는 False)

```python
## [Data Frame] 생성: index 지정 없이
import pandas as pd

df = pd.DataFrame({ 'Name' : ['Kims', 'Lees', 'Parks'],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]},)  #index 강제 부여
df
```

```python
# 'Name' 열을 인덱스로 설정: 'Name'열을 제거하지 않음(drop=False)
ddf = df.set_index('Name')  #'Name'으로 인덱스 대체
ddf
```

```python
# 'Name' 열을 인덱스로 설정: 'Name'열을 제거하지 않음(drop=False)
ddf = df.set_index('Name', drop=False)  #'Name'으로 인덱스 대체
ddf
```

```python
# 'Name' 열을 인덱스로 설정: 'Name'열을 제거하지 않음(drop=False)
ddf = df.set_index('Name', drop=True)  #'Name'으로 인덱스 대체
ddf
```

```python
# 'Name' 열을 인덱스로 설정: 기존 'Name'열은 제거(drop=True)
ddf = df.set_index('Name', drop=True)  #'Name'으로 인덱스 대체
df    #df는 안 바뀜
```

##### inplace=True: 자신을 변경

```python
# 'Name' 열을 인덱스로 설정: 기존 'Name'열은 제거(drop=True)
df.set_index('Name', drop=True, inplace=False)  #'Name'으로 인덱스 대체
df
```

```python
# 'Name' 열을 인덱스로 설정: 기존 'Name'열은 제거(drop=True)
df.set_index('Name', drop=True, inplace=True)  #'Name'으로 인덱스 대체
df
```

### **.reset_index()**
##### 설정된 index를 삭제하고, 기본 숫자 index 부여
###### > drop : True이면 제거되는 index를 열로 추가하지 않음(제거 효과) (default는 True)
###### > drop : False이면 제거되는 index를 열로 추가
###### > inplace : False이면 자신은 안 바꾸고 변경 결과만 반환 (default는 False)

```python
## [Data Frame] 생성: index 있이
import pandas as pd

df = pd.DataFrame({ 'Class' : [1, 1, 2],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]},
                   index = ['Kims', 'Lees', 'Parks'])  #index 강제 부여
df
```

```python
# 현재 인덱스를 일반 열로 변경: index를 reset하고, 열로 추가(drop=False)
ddf = df.reset_index()  #제거되는 index를 열로 추가 (default: drop=Falsee)
ddf
```

### **.reindex()**
##### : 데이터프레임이나 시리즈에서 인덱스나 컬럼 순서를 재정렬하거나 새로 지정할 때 사용
###### > new_index: 새로 지정할 인덱스 리스트
###### > index : 행 인덱스 재정렬
###### > columns : 열 이름 재정렬
###### > fill_value : NaN 대신 채울 값
###### > method : 보간 방식 (ffill, bfill 등)

```python
## [Series] .reindex()로 행 순서 바꾸기
import pandas as pd

s = pd.Series([10, 20, 30], index=["a", "b", "c"])

# 인덱스 순서 바꾸기: + 없는 인덱스 추가
s2 = s.reindex(["c", "b", "a"])
print(s2)
```

```python
## [Series] .reindex()로 행 재구성하기
import pandas as pd

s = pd.Series([10, 20, 30], index=["a", "b", "c"])

# 인덱스 순서 바꾸기: + 없는 인덱스 추가
s2 = s.reindex(["c", "b", "d"])
print(s2)
```

```python
## [Dataframe] 인덱스와 컬럼 순서 모두 변경
df = pd.DataFrame({
    '국어': [90, 80, 70],
    '영어': [85, 95, 75]
}, index=["철수", "영희", "민수"])

# 인덱스와 컬럼 순서 모두 변경
df2 = df.reindex(index=["영희", "민수", "철수"],
                 columns=["영어", "국어"])
print(df2)
```

#### **(알아두면 좋아요) 리스트, 시리즈, 데이터 프레임 복사**
##### > 얕은 복사(shallow copy)와 깊은 복사(deep copy)가 있다.
##### > 얕은 복사는 복사 대상을 공유하는 형태로 참조 방식
##### > 깊은 복사는 복사 대상과 똑같은 데이터를 생성하여 항목을 복사하는 방식

```python
## [Data Frame] 생성
import pandas as pd

df = pd.DataFrame({ 'Name' : ['Kims', 'Lees', 'Parks'],
                   'Class' : [1, 1, 2],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]})  #index 자동 부여
df
```

```python
## 얕은 복사(shallow copy)
df_new = df      # df_new가 df를 참조하는 형태로 복사
df['Mat'] = 100
df_new
```

```python
## 깊은 복사(deep copy)
df_new = df[:]      # df를 슬라이싱 해서 df_new로 복사하므로 두 프레임은 별개
df['Mat'] = 0
df_new
```

```python
## 깊은 복사(deep copy)
df_new = df.copy()
df_new['Mat'] = 90
df_new
```

### **데이터프레임 개별 항목 값 조작**
##### > 반복문을 사용하여 개별 조작
##### > .at[***row_index, col_index***] = ***value***

```python
## [Data Frame] 생성
import pandas as pd

df = pd.DataFrame({ 'Name' : ['Kims', 'Lees', 'Parks'],
                   'Class' : [1, 1, 2],
                   'Eng' : [87, 79, 80], 
                   'Mat' : [80, 90, 80]},)  #index 강제 부여
df
```

```python
## [Data Frame] 조작 : 반복문 사용 접근
for index, row in df.iterrows():
    print(index, row['Name'])  # .at[]은 행과 열로 특정값 검색
```

```python
## [Data Frame] 조작 : 반복문 사용 접근
for index, row in df.iterrows():
    print(index, df.at[index, 'Name'])  # .at[]은 행과 열로 특정값 검색
```

```python
## [Data Frame] 조작 : 반복문 사용 조작
for index, row in df.iterrows():
    if df.at[index, 'Mat'] <= 90:
        df.at[index, 'Mat'] = df.at[index, 'Mat'] + 10
df
```

```python
## [Data Frame] 조작 : 열 값 일괄 적용 조작
df['Mat'] = df['Mat'] - 10
df
```
